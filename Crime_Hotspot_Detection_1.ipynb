# Exploratory Data Analysis(EDA)

```python
#calculating the number of missing values for each column
missing_values = data.isnull().sum()

#plotting histogram for missing values
plt.figure(figsize=(9,5))
plt.barh(missing_values.index, missing_values.values, color='skyblue')
plt.xlabel('number of missing values', fontsize=12)
plt.ylabel('columns', fontsize=12)
plt.title('missing values in each column', fontsize=15)
plt.show()
```

```python
# List of columns to drop
columns_to_drop = ['Mocodes', 'Part 1-2', 'Crm Cd 2','Crm Cd 1', 'Crm Cd 3', 'Crm Cd 4', 'Cross Street', 'Rpt Dist No']

# Dropping the columns
data = data.drop(columns=columns_to_drop, axis=1)

# verifying the remaining columns
print("Remaining columns:", data.columns)
```

```python
from sklearn.impute import SimpleImputer # for filling the missing values in data_set
```

```python
# Columns with missing values
categorical_columns = ['Vict Sex', 'Vict Descent', 'Premis Desc', 'Weapon Desc']
numerical_columns = ['Premis Cd', 'Weapon Used Cd']

# Define imputers
categorical_imputer = SimpleImputer(strategy='most_frequent')  # For categorical columns
numerical_imputer = SimpleImputer(strategy='median')  # For numerical columns
```

```python
# Filling missing values in categorical columns
data[categorical_columns] = categorical_imputer.fit_transform(data[categorical_columns])

# Filling missing values in numerical columns
data[numerical_columns] = numerical_imputer.fit_transform(data[numerical_columns])
```

```python
data.isnull().sum() # checking missing values
```

```python
# Count the number of crimes per area
crimes_per_area = data['AREA NAME'].value_counts()

# Plot the data
plt.figure(figsize=(10, 6))
crimes_per_area.plot(kind='bar', color='skyblue')
plt.title('Number of Crimes per Area', fontsize=14)
plt.xlabel('Area Name', fontsize=12)
plt.ylabel('Number of Crimes', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# IMPORTANT LIBARIRES FOR MACHINE LERANING MODELS

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
# Handling categorical variables using Label Encoding
label_encoder = LabelEncoder()
data['Vict Sex'] = label_encoder.fit_transform(data['Vict Sex'])
data['Weapon Desc'] = label_encoder.fit_transform(data['Weapon Desc'])
data['Premis Desc'] = label_encoder.fit_transform(data['Premis Desc'])
data['Vict Descent'] = label_encoder.fit_transform(data['Vict Descent'])
```

```python
# Encodeing the target variable AREA NAME (our crime hotspot label)
data['AREA NAME'] = label_encoder.fit_transform(data['AREA NAME'])
```

```python
# Select features and target
features = ['LAT', 'LON', 'Year', 'Month', 'Day', 'Hour', 'Crm Cd', 'Vict Age', 'Vict Sex', 'Vict Descent', 'Weapon Desc', 'Premis Desc']
X = data[features]
y = data['AREA NAME']

```

```python
# Scaleing geographic features (LAT, LON)
scaler = StandardScaler()
X[['LAT', 'LON']] = scaler.fit_transform(X[['LAT', 'LON']])
```

```python
# Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
# Initialize models
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
xgb_model = XGBClassifier(n_estimators=100, random_state=42, eval_metric='mlogloss')
knn_model = KNeighborsClassifier(n_neighbors=5)
```

```python
# Training models
rf_model.fit(X_train, y_train)
xgb_model.fit(X_train, y_train)
knn_model.fit(X_train, y_train)
```

```python
# Makeing predictions
y_pred_rf = rf_model.predict(X_test)
y_pred_xgb = xgb_model.predict(X_test)
y_pred_knn = knn_model.predict(X_test)
```

```python
# Model evaluation
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print("XGBoost Accuracy:", accuracy_score(y_test, y_pred_xgb))
print("KNN Accuracy:", accuracy_score(y_test, y_pred_knn))
```

```python
# Classification reports
print("\nRandom Forest Report:\n", classification_report(y_test, y_pred_rf))
```

```python
print("\nXGBoost Report:\n", classification_report(y_test, y_pred_xgb))
```

```python
print("\nKNN Report:\n", classification_report(y_test, y_pred_knn))
```

```python
# Adding XGBoost predictions to test data for visualization
X_test['Predicted_Area'] = y_pred_xgb  # Using XGBoost predictions
```

```python
# Converting encoded labels back to actual AREA NAME
X_test['Predicted_Area_Name'] = label_encoder.inverse_transform(X_test['Predicted_Area'])
```

```python
# Counting occurrences of predicted areas
area_counts = X_test['Predicted_Area_Name'].value_counts()

# Plot bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x=area_counts.index, y=area_counts.values, palette="coolwarm")
plt.xticks(rotation=45)
plt.xlabel("Predicted Crime Hotspot Area")
plt.ylabel("Number of Crimes")
plt.title("Predicted Crime Hotspots - XGBoost Model")
plt.show()
```
